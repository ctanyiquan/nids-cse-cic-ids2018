# 01: Data Merging

## 1.1: Import Libraries

```python
import glob
import os

import pandas as pd
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
pd.set_option("display.width", None)
```

## 1.2: Load Datasets

```python
# Get path to datasets
datasets = glob.glob("../datasets/raw/*.csv")

# Merge datasets into one DataFrame
df = pd.concat([pd.read_csv(dataset, low_memory=False) for dataset in datasets], ignore_index=True)
```

## 1.3: Dtype Inspection

```python
df.info()
```

```python
# Define columns to skip
columns_to_skip = ["Timestamp", "Label", "Flow ID", "Src IP", "Dst IP"]

# Convert all object columns to numeric columns
for column in df.columns:
    if column not in columns_to_skip:
        df[column] = pd.to_numeric(df[column], errors="coerce")
```

```python
df.info()
```

## 1.4: Downcast Dtypes

```python
for column in df.select_dtypes(include="float64").columns:
    df[column] = pd.to_numeric(df[column], downcast="float")
```

```python
df.info()
```

## 1.5: Save DataFrame to Parquet File

```python
# Create directory if it does not exist
os.makedirs("../datasets/raw", exist_ok=True)

# Shuffle DataFrame rows
df = df.sample(frac=1, random_state=67).reset_index(drop=True)

# Save DataFrame to Parquet File
df.to_parquet("../datasets/raw/cse_cic_ids2018_raw.parquet", index=False)
```
